## Statistical Analysis

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Correcting the column names
data.columns = [col.strip() for col in data.columns]
```

```python
# Selecting key variables for statistical analysis
variables_for_analysis = ['1Gender of respondent', '2 How old are you?', '6 How many times you have made an online purchase in the past 1 year?', '15 What is your preferred payment Option?']
```

```python
# Creating contingency tables and conducting chi-square tests
chi_square_results = {}
```

```python
for i in range(len(variables_for_analysis)):
    for j in range(i+1, len(variables_for_analysis)):
        table = pd.crosstab(data[variables_for_analysis[i]], data[variables_for_analysis[j]])
        chi2, p, dof, expected = chi2_contingency(table)
        chi_square_results[f"{variables_for_analysis[i]} vs {variables_for_analysis[j]}"] = {'Chi-Square Value': chi2, 'p-value': p}
```

```python
chi_square_results_df = pd.DataFrame(chi_square_results).T
chi_square_results_df
```

The statistical analysis using chi-square tests reveals the following:

- **Gender vs. Age Group:** The chi-square value is 4.96 with a p-value of 0.291. This suggests no significant association between gender and age group.

- **Gender vs. Frequency of Online Purchases:** The chi-square value is 29.22 with a p-value of approximately 0.00002, indicating a significant association between gender and the frequency of online purchases.

- **Gender vs. Preferred Payment Option:** The chi-square value is 4.63 with a p-value of 0.098, suggesting a marginal association between gender and preferred payment options.

- **Age Group vs. Frequency of Online Purchases:** The chi-square value is 85.87 with a p-value close to 0, indicating a strong association between age group and frequency of online purchases.

- **Age Group vs. Preferred Payment Option:** The chi-square value is 24.80 with a p-value of approximately 0.0017, suggesting a significant association between age group and preferred payment options.

- **Frequency of Online Purchases vs. Preferred Payment Option:** The chi-square value is 52.03 with a p-value close to 0, indicating a strong association between the frequency of online purchases and preferred payment options.

## Feature Engineering

```python
# Feature Engineering

# 1. Binning the "How many times you have made an online purchase in the past 1 year?" into numerical ranges
purchase_freq_bins = {
    'Less than 10 times': '0-10',
    '11-20 times': '11-20',
    '21-30 times': '21-30',
    '31-40 times': '31-40',
    '41 times and above': '41+'
}
```

```python
data['purchase_freq_range'] = data['6 How many times you have made an online purchase in the past 1 year?'].map(purchase_freq_bins)
```

```python
data.head()
```

```python
# 2. Encoding the "Gender of respondent" into a binary format
gender_encoding = {
    'Male': 1,  # Represent Male with 1
    'Female': 0  # Represent Female with 0
}
```

```python
data['gender_encoded'] = data['1Gender of respondent'].map(gender_encoding)
```

```python
# Displaying the new columns
data[['6 How many times you have made an online purchase in the past 1 year?', 'purchase_freq_range', '1Gender of respondent', 'gender_encoded']].head()
```

The feature engineering steps have resulted in the following new features:

1. **`purchase_freq_range`**: This column bins the frequency of online purchases into numerical ranges (e.g., '0-10', '11-20', '21-30', etc.). It simplifies the analysis of purchase frequency by categorizing the data into broader groups.

2. **`gender_encoded`**: The gender of respondents has been encoded into a binary format, with 'Male' represented by 1 and 'Female' by 0. This encoding is particularly useful for any kind of modeling or correlation analysis that requires numerical input.

These engineered features can now be used in various types of analyses or machine learning models to gain further insights or predict future trends.

```python
# Encoding the categorical features
for column in selected_features.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    selected_features.loc[:, column] = le.fit_transform(selected_features[column])
```

```python
# Making predictions on the test set
y_pred = rf_classifier.predict(X_test)
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.compose import ColumnTransformer
```

```python
# Selecting relevant features
selected_features = data[[
    'gender_encoded', 
    '2 How old are you?', 
    '3 Which city do you shop online from?',
    '5 Since How Long You are Shopping Online ?', 
    '7 How do you access the internet while shopping on-line?',
    '15 What is your preferred payment Option?', 
    '17 Why did you abandon the “Bag”, “Shopping Cart”?'
]]
```

```python
# One-Hot Encoding for categorical features
ct = ColumnTransformer([
        ('onehot', OneHotEncoder(), [
            '2 How old are you?', 
            '3 Which city do you shop online from?',
            '5 Since How Long You are Shopping Online ?', 
            '7 How do you access the internet while shopping on-line?',
            '15 What is your preferred payment Option?', 
            '17 Why did you abandon the “Bag”, “Shopping Cart”?'
            # Add other categorical columns that require one-hot encoding
        ]),
    ], remainder='passthrough')
```

```python
X = ct.fit_transform(selected_features)
y = data['purchase_freq_range']
```

```python
# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Building the Random Forest Classifier with hyperparameter tuning
rf_classifier = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=5)
rf_classifier.fit(X_train, y_train)
```

```python
# Cross-validation
scores = cross_val_score(rf_classifier, X, y, cv=5)
print(f"Cross-validated scores: {scores.mean()}")
```

```python
# Making predictions on the test set
y_pred = rf_classifier.predict(X_test)
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
report = classification_report(y_test, y_pred)
print(report)
```
